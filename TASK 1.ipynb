```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
file_path = r'C:\Users\Raksha\Documents\Notes DA\python notes\DATASET FOR PYTHON\csv _files\datass.csv'
# Load the dataset
dataset = pd.read_csv(file_path)
```

Determine the top three most
common cuisines in the dataset.

```python
print (dataset.shape)
```

```python
print (dataset.columns)
```

```python
print(dataset['Cuisines'])
```

```python
cuisines_count=dataset['Cuisines'].str.split(', ').explode('Cuisines').value_counts()
print(cuisines_count)
top_cuisines=cuisines_count.head(3)
print("The top 3 cuisines are : ", top_cuisines)
```

```python
colours=['green','orange','red']
plt.bar(top_cuisines.index,top_cuisines.values,color=colours)
plt.xlabel('cuisine')
plt.ylabel('count')
plt.title('TOP THREE CUISINES')
plt.show()
```

```python
calculate the percentage of restarents that serve each of top cuisines
```

```python
total_restaurant = len(dataset)
print(total_restaurant)
top_cuisine10=cuisines_count.head()
percentages =(top_cuisine10 / total_restaurant)* 100
print("The Market share of Top 5 Cuisines are:Â ",percentages)
```

```python
plt.bar(top_cuisine10.index,percentages.values,color=colours)
plt.xlabel('cuisine name')
plt.ylabel('percentage')
plt.title('TOP five CUISINES % ')
plt.figure(figsize=(12,6))
plt.show()
```
